```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt 
import time
import random
import math
import time


class Circle:
    
    def __init__(self):
        self.center=(70,70)
        self.direction=(0,0)
        self.speed=(10,10)
        
        self.color=[200,200,200]
        self.radius=70
        self.canvas=np.zeros((500,500,3),dtype='uint8')
       # print('ready')
   
    def createCircle(self,center,radius,color):
        self.center=center
        self.radius=radius 
        self.color=color

    
    def drawCircle(self,canvas):
        cv.circle(canvas,self.center,self.radius,self.color,-1)
        return canvas
    
    def move(self):
       # print('dir=',self.direction)
        x_center=self.center[0]+self.speed[0]*self.direction[0]
        y_center=self.center[1]+self.speed[1]*self.direction[1]
        self.center=(x_center,y_center)
       # print(self.center)
        

print('start')
circle=Circle()
circle.createCircle((50,50),50,[255,255,255])
circle.drawCircle(circle.canvas)
plt.imshow(circle.canvas)
copy=circle.canvas.copy()

low=130
high=370

print('start')
while True:
    print('True')
    flag=True
      
    print('reached')
    base=circle.center[0];
    height=random.randint(low,high)
    hyp=int(math.sqrt((base**2+height**2)))
    print('base and height',base,height)
#----------------------------------------------------------------------------------------    
    for i in range(base,450):       
        c=Circle()
        copy=circle.drawCircle(c.canvas)
        cv.imshow('test',copy)
        key=cv.waitKey(1) & 0xff
        if key==ord('q'):
            cv.destroyAllWindows()
            flag=False
            break
        else:
            circle.center=(i,circle.center[1])
            if i %(max(base//height,height//base))==0:
                circle.center=(circle.center[0],circle.center[1]+1)
            print(circle.center)
         #   time.sleep(0.0005)              
    if flag==False:
        break 
    print('1st round over')
 #--------------------------------------------------------------------------------------
    height=circle.center[1]
    new_height=500-height 
    point=random.randint(low,high)   
    base=500-point
    hyp=int(math.sqrt(base**2+new_height**2))
    
    
    for i in range(height,451,):  
        
        c=Circle()
        copy=circle.drawCircle(c.canvas)
        cv.imshow('test',copy)
        key=cv.waitKey(1) & 0xff
        if key==ord('q'):
            cv.destroyAllWindows()
            flag=False
            break
        else:
            circle.center=(circle.center[0],i)
            if i %(base//height+1)==0:
                circle.center=(circle.center[0]-(base//height)-1,i)
            print(circle.center)
         #   time.sleep(0.005)
    print('outside')            
    if flag==False:
        break 
        
 #--------------------------------------------------------------------------------------  
    height=circle.center[1]
    base=circle.center[0]  
    point=random.randint(low,high)   
    new_height=500-point

    hyp=int(math.sqrt(50**2+new_height**2))
    
    print('3ed start')
    for i in range(base,49,-1):    
        c=Circle()
        copy=circle.drawCircle(c.canvas)
        cv.imshow('test',copy)     
        
        key=cv.waitKey(1) & 0xff
        if key==ord('q'):
            cv.destroyAllWindows()
            flag=False
            break
        else:
            circle.center=(i,circle.center[1])
            if i %(max(base//height,height//base))==0:
                circle.center=(i,circle.center[1]-max(base//height,height//base))
            print(circle.center)
          #  time.sleep(0.005)
    print('outside')            
    if flag==False:
        break 
   # if (cv.waitKey(0) & 0xff)==ord('q'):
    #    cv.destroyAllWindows()
    
#----------------------------------------------------------------------------
    height=circle.center[1]
    base=circle.center[0]  
    point=random.randint(low,high)  
    
    print('4ed start')
    for i in range(height,49,-1):    
        c=Circle()
        copy=circle.drawCircle(c.canvas)
        cv.imshow('test',copy)     
        
        key=cv.waitKey(1) & 0xff
        if key==ord('q'):
            cv.destroyAllWindows()
            flag=False
            break
        else:
            circle.center=(circle.center[0],i)
            if i %(max(base//height,height//base))==0:
                circle.center=(circle.center[0]+max(base//height,height//base),i)
            print(circle.center)
        #    time.sleep(0.005)
    print('outside')            
    if flag==False:
        break 
    #if (cv.waitKey(0) & 0xff)==ord('q'):
     #   cv.destroyAllWindows()
    #break














    

    
    
   
    
    
        
```
